```python
def parse_gla(value):
    if pd.isna(value):
        return np.nan

    value = value.replace(',', '')
    
    if 'SqM' in value or 'sqm' in value.lower():
        try:
            num = float(value.split()[0])
            return num * 10.764
        except:
            return np.nan
    
    value.replace('+/-', '')
    
    if "SqFt" in value or 'sqft' in value.lower() or 'sf' in value.lower():
        try:
            num = ''.join(c for c in value if (c.isdigit() or c == '.'))
            return float(num)
        except:
            return np.nan
        
    try:
        return float(value)
    except:
        return np.nan
    
normalized_df['gla_sqft'] = normalized_df['subject.gla'].apply(parse_gla)
```

```python
def parse_num_beds(value):
    if pd.isna(value):
        return np.nan
    value = str(value).strip()
    
    if '+' in value:
        try:
            return sum(int(part) for part in val.split('+'))
        except:
            return np.nan
        
    try:
        return int(value)
    except:
        return np.nan
    
normalized_df['beds_num'] = normalized_df['subject.num_beds'].apply(parse_num_beds)
```

```python
import re
```

```python
def parse_baths_num(value):
    if pd.isna(value):
        return np.nan

    value = str(value).strip().upper()
    
    # Format: "2:1" → 2 full, 1 half
    if ':' in value:
        try:
            full, half = map(int, value.split(':'))
            return full + 0.5 * half
        except:
            return np.nan

    # Format: "2F 1H" or "1F 1H"
    match = re.match(r'(?:(\d+)F)?\s*(?:(\d+)H)?', value)
    if match:
        full = int(match.group(1)) if match.group(1) else 0
        half = int(match.group(2)) if match.group(2) else 0
        return full + 0.5 * half

    # Format: "2F", "1F", "3F" (only full baths)
    if value.endswith('F') and value[:-1].isdigit():
        return int(value[:-1])

    # Format: "2", "3" (assume full baths only)
    if value.isdigit():
        return int(value)

    # Format: "2 Full/1Half"
    match_alt = re.match(r'(\d+)\s*FULL\s*/\s*(\d+)HALF', value)
    if match_alt:
        return int(match_alt.group(1)) + 0.5 * int(match_alt.group(2))

    return np.nan

normalized_df['baths_num'] = normalized_df['subject.num_baths'].apply(parse_baths_num)
```

```python
def parse_lot_size(value):
    if pd.isna(value):
        return np.nan

    value = str(value).strip().lower().replace(',', '').replace('+/-', '')

    # Handle acres (e.g., "1.25 acres", "0.5ac")
    if 'acre' in value or 'ac' in value:
        try:
            number = float(re.findall(r'[\d.]+', value)[0])
            return number * 43560
        except:
            return np.nan

    # Handle square meters (e.g., "549 sqm")
    if 'sqm' in value or 'sq m' in value:
        try:
            number = float(re.findall(r'[\d.]+', value)[0])
            return number * 10.7639
        except:
            return np.nan

    # Handle square feet (e.g., "10570 sqft", "10570 sq ft")
    if 'sqft' in value or 'sq ft' in value:
        try:
            return float(re.findall(r'[\d.]+', value)[0])
        except:
            return np.nan

    # Handle numbers without units (assume they're in square feet)
    try:
        return float(re.findall(r'[\d.]+', value)[0])
    except:
        return np.nan
    
normalized_df['lot_size_sqft'] = normalized_df['subject.lot_size_sf'].apply(parse_lot_size)
```

```python
def parse_year_to_age(value):
    if pd.isna(value):
        return np.nan
    
    value = str(value).strip().lower()
    
    if 'new' in value:
        return datetime.now().year - 1
    
    match = re.search(r'\d{4}', value)
    if match:
        return int(match.group(0))
    
    return np.nan

normalized_df['parsed_year_built'] = normalized_df['subject.year_built'].apply(parse_year_to_age)

current_year = datetime.now().year
normalized_df['property_age'] = current_year - normalized_df['parsed_year_built']
```

```python
def parse_gla(val):
    if pd.isnull(val):
        return None
    val = str(val).lower().strip()
    # Handle common square footage patterns
    match = re.search(r"[\d,\.]+", val)
    if match:
        try:
            # Remove commas and convert to float
            return float(match.group(0).replace(",", ""))
        except ValueError:
            return None
    return None
```

```python
def parse_comp_age(value):
    if pd.isnull(value):
        return np.nan

    value_str = str(value)

    # If the value looks like a year (e.g. '2015 +/-')
    year_match = re.search(r"(19\d{2}|20\d{2})", value_str)
    if year_match:
        year = int(year_match.group(1))
        if 1800 <= year <= datetime.now().year:
            return datetime.now().year - year

    # Else look for first standalone number (age)
    num_match = re.search(r"\d{1,3}", value_str)  # allows up to 999 years old
    if num_match:
        return int(num_match.group(0))

    return np.nan
```

```python
def run_kmeans_for_group(df):
    from sklearn.cluster import KMeans
    from sklearn.preprocessing import StandardScaler

    features = [
        "distance_to_subject_km", "parse_gla", "parse_lot_size_sf",
        "parsed_beds", "parsed_full_baths", "total_baths",
        "parsed_year_built", "property_age"
    ]

    # Drop rows with missing features
    df_clean = df.dropna(subset=features)

    if len(df_clean) < 2:
        return pd.Series([-1] * len(df), index=df.index)  # not enough for clustering

    # Scale features
    X = StandardScaler().fit_transform(df_clean[features])
    kmeans = KMeans(n_clusters=2, n_init=10, random_state=42)
    labels = kmeans.fit_predict(X)

    # Reconstruct full output aligned to original group_df
    out = pd.Series(-1, index=df.index)
    out.loc[df_clean.index] = labels
    return out

# Apply to each group and assign
properties_df["cluster"] = properties_df.groupby("group_id", group_keys=False).apply(run_kmeans_for_group)
```

```python
from sklearn.cluster import KMeans

def run_kmeans_for_group(group_df):
    if len(group_df) < 2:
        return [-1] * len(group_df)

    # You can adjust features as needed
    features = group_df[["parse_gla", "parsed_beds"]]

    km = KMeans(n_clusters=2, random_state=42)
    result = km.fit_predict(features)

    # Debugging safety check
    assert len(result) == len(group_df), f"Length mismatch: got {len(result)} vs expected {len(group_df)}"

    return result
```

```python
from scipy.spatial.distance import euclidean


def find_closest_cluster(subject_row, group_clusters):
    # Convert subject row to array
    subject_array = subject_row.to_numpy()

    # Check for NaNs
    if np.isnan(subject_array).any() or group_clusters.isnull().values.any():
        return None

    # Calculate distances
    distances = group_clusters.apply(lambda row: euclidean(row.to_numpy(), subject_array), axis=1)

    # Get index safely
    min_idx = distances.idxmin()
    return min_idx[1] if isinstance(min_idx, tuple) else min_idx

```

```python
precision_list, recall_list, f1_list = [], [], []

for group in actual_comps['orderID']:
    true_set = actual_comps.loc[actual_comps['orderID'] == group, 'comp_ids'].values[0]
    pred_set = predicted_sets.get(group, set())

    tp = len(true_set & pred_set)
    fp = len(pred_set - true_set)
    fn = len(true_set - pred_set)

    precision = tp / (tp + fp) if (tp + fp) else 0
    recall = tp / (tp + fn) if (tp + fn) else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)

avg_precision = sum(precision_list) / len(precision_list)
avg_recall = sum(recall_list) / len(recall_list)
avg_f1 = sum(f1_list) / len(f1_list)

print(f"Precision: {avg_precision:.3f}, Recall: {avg_recall:.3f}, F1 Score: {avg_f1:.3f}")
```
